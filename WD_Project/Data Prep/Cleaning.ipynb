```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
m_path = "../Data/meteorite-landings.csv"
e_path = "../Data/EQ_database.csv"
v_path = "../Data/volerup.csv"
meteor_df = pd.read_csv(m_path, encoding="utf-8")
eq_df = pd.read_csv(e_path, encoding="utf-8")
volcano_df = pd.read_csv(v_path, encoding="utf-8")
meteor_df.head()
```

```python
volcano_df.head()
```

```python
eq_df.head()
```

```python
# Cleaning Meteor Dataset
list(meteor_df)
```

```python
meteor_df = meteor_df.drop(['id', 'nametype', 'recclass', 'fall', 'GeoLocation'], axis=1)
```

```python
meteor_df.columns = ['Name', 'Mass', 'Year', 'Latitude', 'Longitude']
meteor_df.dropna()
meteor_df.head()
```

```python
#Labeling Metoer events
meteor_passer= []
for row in meteor_df.index:
    if row >= 0:
        meteor_passer.append('Meteor')       
```

```python
meteor_df['Type'] = meteor_passer
meteor_df.head()
```

```python
#Reapplying NA where data is missing
meteor_df.fillna(str("NaN"))
```

```python
#Dropping NA
meteor_df.dropna()
```

```python
#Sorting by Mass
meteor_df = meteor_df.sort_values(by='Mass', ascending=False)
meteor_df = meteor_df.reset_index(drop=True)
meteor_df.head()
```

```python
#Assigning Severity Rating
mass_passer= []
for row in meteor_df['Mass']:
    if row >= 10000000.0:
        mass_passer.append('5')
    elif row >= 4300000.0:
        mass_passer.append('4')
    elif row >= 3000000.0:
        mass_passer.append('3')
    elif row >= 600000.0:
        mass_passer.append('2')
    else:
        mass_passer.append('1')
```

```python
#Dropping the rating into the DataFrame
meteor_df['Rating'] = mass_passer
meteor_df.head()
```

```python
# Cleaning Earthquake Dataset
list(eq_df)
```

```python
#Dropping columns to match with others
eq_df = eq_df.drop(['Type.1', 'Depth', 'Depth Error', 'Depth Seismic Stations', 'Magnitude Seismic Stations','Azimuthal Gap', 'Horizontal Distance', 'Horizontal Error', 'Root Mean Square', 'ID', 'Source', 'Location Source', 'Magnitude Source', 'Status', 'Magnitude Type','Magnitude Error' ], axis=1)
eq_df.head()
```

```python
eq_df = eq_df.sort_values(by='Magnitude', ascending=False)
eq_df = eq_df.reset_index(drop=True)
eq_df.head(50)
```

```python
#Normalizing the dates
eq_df['Date'] = eq_df['Date'].str[-4:]
```

```python
#Normalizing the dates
eq_df = eq_df.rename(columns={
    'Date':'Year'
})
```

```python
#Assigning severity rating
mag_passer= []
for row in eq_df['Magnitude']:
    if row >= 8.5:
        mag_passer.append('5')
    elif row >= 8.0:
        mag_passer.append('4')
    elif row >= 7.7:
        mag_passer.append('3')
    elif row >= 7.3:
        mag_passer.append('2')
    else:
        mag_passer.append('1')
```

```python
volcano_df
```

```python
eq_df['Rating']=mag_passer
eq_df.head()
```

```python
# Cleaning Volcano Dataset
list(volcano_df)
```

```python
volcano_df = volcano_df.drop(['Month', 'Day', 'TSU', 'EQ', 'Name','Location', 'Elevation', 'Agent', 'DEATHS_DESCRIPTION', 'MISSING', 'MISSING_DESCRIPTION', 'INJURIES', 'INJURIES_DESCRIPTION', 'DAMAGE_MILLIONS_DOLLARS', 'DAMAGE_DESCRIPTION','HOUSES_DESTROYED', 'HOUSES_DESTROYED_DESCRIPTION', 'DAMAGE_MILLIONS_DOLLARS', 'DAMAGE_DESCRIPTION', 'TOTAL_DEATHS_DESCRIPTION', 'TOTAL_MISSING', 'TOTAL_MISSING_DESCRIPTION', 'TOTAL_INJURIES', 'TOTAL_INJURIES_DESCRIPTION', 'TOTAL_DAMAGE_MILLIONS_DOLLARS', 'TOTAL_DAMAGE_DESCRIPTION', 'TOTAL_HOUSES_DESTROYED', 'TOTAL_HOUSES_DESTROYED_DESCRIPTION','TOTAL_DEATHS','Time', 'Status', 'DEATHS'], axis=1)
volcano_df.head()
```

```python
volcano_df.fillna(str("NaN"))
```

```python
volcano_df.dropna()
```

```python
#Sorting by VEI
volcano_df = volcano_df.sort_values(by='VEI', ascending=False)
volcano_df = volcano_df.reset_index(drop=True)
volcano_df.head(100)
```

```python
#Assing severity rating
volc_rating= []
for row in volcano_df['VEI']:
    if row >= 6:
        volc_rating.append('5')
    elif row >= 5:
        volc_rating.append('4')
    elif row >= 4:
        volc_rating.append('3')
    elif row >= 3:
        volc_rating.append('2')
    else:
        volc_rating.append('1')
```

```python
volcano_df['Rating'] = volc_rating
```

```python
#Duplicative name, renaming to prep to place the"Type" field to match the others
volcano_df['Volcano Type'] = volcano_df['Type']
```

```python
volc_passer= []
for row in volcano_df.index:
    if row >= 0:
        volc_passer.append('Volcano')  
```

```python
volcano_df['Type'] = volc_passer
```
